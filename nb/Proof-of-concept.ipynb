# Proof-of-concept 

```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.decomposition import PCA

data_m, target_v = load_iris(return_X_y=True)
data_df = pd.DataFrame(data_m)

pca = PCA(n_components=4)
pca.fit(data_df)

loadings_df = pd.DataFrame(pca.components_)
loadings_df.index = [f"pc: {i} ({round(v, 2)})" 
                     for i, v in 
                     enumerate(pca.explained_variance_ratio_)]
loadings_df.plot(kind="bar", rot=0, figsize=(12,5))
```
